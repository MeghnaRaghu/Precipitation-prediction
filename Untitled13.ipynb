```python
from netCDF4 import Dataset

# File paths
file_paths = [
    "/content/drive/MyDrive/gridded nc4/RF25_ind2020_rfp25.nc",
    "/content/drive/MyDrive/gridded nc4/RF25_ind2021_rfp25 (1).nc",
    "/content/drive/MyDrive/gridded nc4/RF25_ind2022_rfp25.nc"
]

# Load the NetCDF files and read rainfall data
rainfall_data = []
for file_path in file_paths:
    with Dataset(file_path, 'r') as dataset:
        rainfall_data.append(dataset.variables['RAINFALL'][:])

# Now you have the rainfall data for each year stored in separate arrays
# You can perform analysis or visualization as needed
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Aggregate daily rainfall data to get annual rainfall
annual_rainfall = [np.sum(year_data, axis=0) for year_data in rainfall_data]

# Plotting aggregated annual rainfall for each year
years = [2020, 2021, 2022]
for i, year_rainfall in enumerate(annual_rainfall):
    plt.figure(figsize=(10, 6))
    plt.imshow(year_rainfall, cmap='Blues', origin='lower', extent=[0, 1, 0, 1])
    plt.colorbar(label='Annual Rainfall (mm)')
    plt.title(f"Annual Rainfall for {years[i]}")
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(years, annual_rainfall, test_size=0.2, random_state=42)

# Train Random Forest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print evaluation metrics
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```
